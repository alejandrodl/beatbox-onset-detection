```python
import os
import numpy as np
```

```python
modes = ['eng_mfcc_env','eng_all','syllall','E2E']
clfs = ['knn3','knn5','knn7','knn9','knn11','logr','rf','xgboost']

# Calculate boxeme-wise weights

num_test_boxemes = []
for part in range(28):
    test_dataset = np.load('../../data/interim/AVP_1024/Dataset_Test_' + str(part).zfill(2) + '.npy')
    num_test_boxemes.append(test_dataset.shape[0])
boxeme_wise_weights = num_test_boxemes/np.sum(np.array(num_test_boxemes))

for mode in modes:
    
    if mode!='E2E':

        # Results participant-wise

        print('\n')
        print('Participant-wise')
        print('\n')

        for b in range(len(clfs)):
            clf = clfs[b]
            accuracies_raw = np.load('../../results/' + mode + '/accuracies.npy')[:,:,b,:]
            accuracies_mean = 0
            accuracies_std = 0
            for n in range(28):
                accuracies_mean += np.mean(accuracies_raw[:,:,n])
                accuracies_std += np.std(accuracies_raw[:,:,n])
            accuracies_mean /= 28
            accuracies_std /= 28
            print([mode,clf])
            print([accuracies_mean,accuracies_std])

        # Results boxeme-wise

        print('\n')
        print('Boxeme-wise')
        print('\n')

        for b in range(len(clfs)):
            clf = clfs[b]
            accuracies_raw = np.load('../../results/' + mode + '/accuracies.npy')[:,:,b,:]
            for i in range(accuracies_raw.shape[0]):
                for j in range(accuracies_raw.shape[1]):
                    accuracies_raw[i,j] *= boxeme_wise_weights*28
            accuracies_mean = 0
            accuracies_std = 0
            for n in range(28):
                accuracies_mean += np.mean(accuracies_raw[:,:,n])
                accuracies_std += np.std(accuracies_raw[:,:,n])
            accuracies_mean /= 28
            accuracies_std /= 28
            print([mode,clf])
            print([accuracies_mean,accuracies_std])
            
    else:
        
        # Results participant-wise

        print('\n')
        print('Participant-wise')
        print('\n')
        
        accuracies_raw = np.load('../../results/' + mode + '/accuracies.npy')
        accuracies_mean = 0
        accuracies_std = 0
        for n in range(28):
            accuracies_mean += np.mean(accuracies_raw[n])
            accuracies_std += np.std(accuracies_raw[n])
        accuracies_mean /= 28
        accuracies_std /= 28
        print([mode])
        print([accuracies_mean,accuracies_std])

        # Results boxeme-wise

        print('\n')
        print('Boxeme-wise')
        print('\n')

        accuracies_raw = np.load('../../results/' + mode + '/accuracies.npy')
        for i in range(accuracies_raw.shape[1]):
            for j in range(accuracies_raw.shape[2]):
                accuracies_raw[:,i,j] *= boxeme_wise_weights*28
        accuracies_mean = 0
        accuracies_std = 0
        for n in range(28):
            accuracies_mean += np.mean(accuracies_raw[n])
            accuracies_std += np.std(accuracies_raw[n])
        accuracies_mean /= 28
        accuracies_std /= 28
        print([mode])
        print([accuracies_mean,accuracies_std])
```
